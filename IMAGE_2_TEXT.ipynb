```python
from PIL import Image
from PIL import ImageFilter
from PIL import ImageEnhance
import numpy
import pytesseract
from pytesseract import Output
import cv2
import matplotlib.pyplot as plt
```

```python
IM = 'SS_01.png'
image = Image.open(IM)

plt.imshow(image)
```

```python
source = IM

#source = image

img = cv2.imread(source)
text=pytesseract.image_to_string(img)
if len(text) > 3:
    print(text)
else:
    print("Texto no reconhecido")
```

## Anlise de imagens com filtros diferentes do Pillow

```python
#imagem = cv2.imread(IM)

imagem = Image.open(IM)

#imagem_pb = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

imagem_pb = imagem.convert('LA')

plt.imshow(imagem_pb)
```

```python
#source = "imagem_pb"

#img = cv2.imread(source)
text=pytesseract.image_to_string(imagem_pb)
if len(text) > 3:
    print(text)
else:
    print("Texto no reconhecido")
```

```python
# imagem filtrado com nitidez

imagem = Image.open(IM)

imagem_f = imagem.filter(ImageFilter.SHARPEN)

plt.imshow(imagem_f)
```

```python
text=pytesseract.image_to_string(imagem_f)
if len(text) > 3:
    print(text)
else:
    print("Texto no reconhecido")
```

## Anlise da imagem em RGB

```python
imagem = cv2.imread(IM)
imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem_rgb)
```

```python
#source = "imagem_pb"

#img = cv2.imread(source)
text=pytesseract.image_to_string(imagem_rgb)
if len(text) > 3:
    print(text)
else:
    print("Texto no reconhecido")
```

```python
imagem = cv2.imread(IM)

imagem_gr = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

imagem_invertida = cv2.bitwise_not(imagem_gr)

plt.imshow(imagem_invertida )
```

```python
#source = "imagem_pb"

#img = cv2.imread(source)
text=pytesseract.image_to_string(imagem_invertida)
if len(text) > 3:
    print(text)
else:
    print("Texto no reconhecido")
```
